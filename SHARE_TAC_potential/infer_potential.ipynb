# output cross_attention.csv

##### loading Mask
The dimensions of the original cross-attention matrix are n (gene) x m (peak). To accelerate the computation and reduce memory usage, we load the genomic distance file and mupltiply it with cross-attention matrix.

* share_matrix.mtx：genomic distance within 1200kb ("window" in GLUE experiment)
* share_gene.list: ordered genes list in window.mtx
* share_peak.list: ordered peaks list in window.mtx
* result_crossattention.npz: cross-attention matrix generated by Attune

```python
window = sc.read('./share_matrix.mtx')
wgene = pd.read_csv('./share_gene.list',header=None)
wpeak = pd.read_csv('./share_peak.list',header=None)
```

```python
f = np.load('./result/cross_attention/result_crossattention.npz')
RNA_id = f['RNA_id']
cross_attention_1 = f['cross_attention_1']
cross_attention_2 = f['cross_attention_2']
cross_attention_1_array = np.vstack(cross_attention_1)
cross_attention_2_array = np.vstack(cross_attention_2)
cross_attention = cross_attention_2_array.T @ cross_attention_1_array
cross_attention_wocls = cross_attention[1:,1:]
cross_attention_wocls.shape
```

##### Mask and output
* adata_rna_SHARE_TAC.h5ad: preprocess rna file
* adata_atac_SHARE_TAC.h5ad: preprocess atac file

```python
adata_rna = sc.read('./data/adata_rna_SHARE_TAC.h5ad')
adata_rna_1 = adata_rna[RNA_id]
adata_atac = sc.read('./data/adata_atac_SHARE_TAC.h5ad')
adata_atac_1 = adata_atac[RNA_id]
```

```python
df = pd.DataFrame(cross_attention_wocls, columns=adata_atac_1.var_names.map(lambda x: x.replace('-', ':', 1)), index=adata_rna_1.var_names)
df = df.loc[wgene[0].values,:]
df = df.loc[:,wpeak[0].values]
df2 = window.X.multiply(df.values)
df2 = pd.DataFrame.sparse.from_spmatrix(df2)
df2.columns = df.columns.values
df2['gene'] = df.index.values
df_new = pd.melt(df2, id_vars='gene',value_vars=df2.columns.values[:-1])
df_new_f = df_new[df_new.value>0]
df_new_f.to_csv(f'./result/cross_attention/cross_attention.csv', index=False)
```
